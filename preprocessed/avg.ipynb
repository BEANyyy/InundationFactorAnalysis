배수펌프장 ( 날리기

```python
import pandas as pd

pump = pd.read_csv("한국국토정보공사_배수펌프장 서비스.csv")
avlty = pump['item/avilaty']
avlty = avlty.to_frame()
avlty['result'] = avlty['item/avilaty'].str.split('(').str[0]
avlty['item/avilaty'] = avlty['result']
avlty = avlty.drop(['result'], axis=1)
avlty
```

원래 데이터랑 합치기

```python
pump['item/avilaty'] = avlty
pump.to_csv("한국국토정보공사_배수펌프장 서비스_수정.csv", mode='w')
```
